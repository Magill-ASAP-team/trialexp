```python
import neo
from elephant import statistics, kernels
import quantities as pq
import matplotlib.pylab as plt
```

```python
sigma_ms = 200 # ms for half-gaussian kernel size (1SD duration)
kernel = kernels.ExponentialKernel(sigma=sigma_ms*pq.ms)
spiketrain = neo.SpikeTrain([1], t_start=-1, t_stop=5, units='s')
counts = statistics.time_histogram([spiketrain], 10*pq.ms)

# kernel_pdf = kernel(spiketrain)
inst_rates = statistics.instantaneous_rate(
                    spiketrains= [spiketrain],
                    sampling_period = 10 * pq.ms, 
                    kernel=kernel, 
                    cutoff=5.0, 
                    trim=False, 
                    center_kernel=True, 
                    border_correction=False)
```

```python
plt.plot(inst_rates.times, inst_rates.T[0])
```
